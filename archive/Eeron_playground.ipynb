# Vader

```python
from nltk import tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
train.comment_text[:10]
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
comment = train.comment_text[0]
comment
```

```python
sid.polarity_scores(comment)
```

```python
%%time
neg = train.comment_text[:10000].apply(lambda x: sid.polarity_scores(x)['neg'])
```

```python
%%time
pos = train.comment_text[:10000].apply(lambda x: sid.polarity_scores(x)['pos'])
```

```python
%%time
neu = train.comment_text[:10000].apply(lambda x: sid.polarity_scores(x)['neu'])
```

```python
train10k = train[:10000].copy()
train10k['vader_neg'] = neg
train10k['vader_pos'] = pos
train10k['vader_neu'] = neu
```

```python
train10k.groupby('obscene').vader_pos.mean()
```

```python
train10k.groupby('severe_toxic').vader_neg.mean()
```

```python
train10k.vader_neg.mean()
```

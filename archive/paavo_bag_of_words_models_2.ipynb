```python
import pandas as pd
```

```python
dataset = pd.read_csv('data/train.csv').set_index('id')
```

```python
documents = dataset.comment_text
documents.head()
```

## new bag

```python
from sklearn.model_selection import train_test_split
```

```python
predict_columns = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

X, y = dataset.comment_text, dataset[predict_columns]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

for s in X_train, X_test, y_train, y_test:
    print(s.shape)
```

```python
from nltk import word_tokenize
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import LinearSVC
```

```python
our_column = 'toxic'
```

```python
%%time
clf = Pipeline([
    ('vectorizer', CountVectorizer(analyzer="word",
                                   tokenizer=word_tokenize,
                                   max_df=0.5) ),
    ('classifier', LinearSVC())
])
clf.fit(X_train, y_train[our_column])
```

```python
%%time
score = clf.score(X_test, y_test[our_column])
```

```python
score
```

```python
%%time
predict_test = clf.predict(X_test)
```

```python
predict_test
```

```python
predict_df = pd.DataFrame(y_test)
predict_df['predicted_' + our_column] = predict_test
predict_df.head()
```

```python
cutoff = 0.04
tempdf = predict_df[['toxic', 'predicted_toxic']].copy()
tempdf.predicted_toxic = tempdf.predicted_toxic.apply(lambda x: 1 - cutoff if x == 1 else cutoff)
tempdf
```

```python
tempdf.to_csv('partial_results/predict_df.csv')
```

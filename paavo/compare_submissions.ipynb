```python
import pandas as pd
```

```python
sub1 = pd.read_csv('./nb_svm_sklearny_submission.csv')
sub2 = pd.read_csv('./nb_svm_strong_linear_baseline_submission.csv')
```

```python
sub1.equals(sub2)
```

```python
merged = pd.merge(sub1, sub2, on='id')
merged.shape
```

```python
merged.head(10)
```

```python
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

for label in label_cols: 
    comp = merged[[label + '_x', label + '_y']]
    ssd = ((comp.iloc[:, 0] - comp.iloc[:, 1])**2).sum()
    print('{} sum square difference: {}'.format(label, ssd))
```
